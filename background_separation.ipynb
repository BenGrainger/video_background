```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
video = r'C:\Users\BMLab21\Documents\CrabStreams\2022-03-30.avi'
```

```python
cap = cv2.VideoCapture(video)

bg_subtractor = cv2.createBackgroundSubtractorMOG2() # create background subtractor instance
history = 1000 # history is number of frames to be used
Learning_rate = 1.0/history 
```

```python
while(True):
    ret, frame = cap.read()
    if ret == True:
        img = np.asarray(frame)
        mask = bg_subtractor.apply(frame, learningRate=Learning_rate) 
        mask = cv2.cvtColor(mask, cv2.COLOR_GRAY2BGR)
    else:
        break

cap.release()
cv2.destroyAllWindows()
```
